```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('tienda_ventas.csv')
df
```

```python
df["mes"] = df['date'].apply(lambda x: x[:7])
df
```

```python
ventasTiendasMes = df.groupby(['mes','store_nbr']).agg({
    'sales': 'sum'
})
ventasTiendasMes
```

```python
ventasTiendasMes = ventasTiendasMes.pivot_table(index='mes', columns='store_nbr', values='sales')
ventasTiendasMes
```

```python
fig = px.line(ventasTiendasMes)
fig.show()
```
