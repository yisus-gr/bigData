```python
import pandas as pd
```

```python
df = pd.read_csv('viajes_surfistas.csv')
df
```

```python
df["paisOrigen"] = df["homecountry"].apply(
    lambda x: x if x == "United States of America" or x == "Japan" or x == "Australia" or x == "Spain" else "Otros")
df
```

```python
paises = df.groupby('paisOrigen').count()
paises
```

```python
paises = paises.sort_values(by='homecountry') #ordenamos los paises por cantidad de participantes
paises
```

```python
paises.plot.pie(y="homename", startangle=90, autopct='%1.1f%%', )
```
